```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from scipy import stats
from scipy.stats import norm,skew

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns',100)
```

```python
df_train = pd.read_csv('train.csv')

df_train.head()
```

```python
df_test = pd.read_csv('test.csv')

df_test.head()
```

## 1. SalePrice

```python
(mu,sigma) = norm.fit(df_train['SalePrice'])
print('mu = {:.2f} and sigma = {:.2f}'.format(mu,sigma))

sns.distplot(df_train['SalePrice'],fit = norm);
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f})'.format(mu,sigma)],loc = 'best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution');
```

```python
df_train['SalePrice'] = np.log1p(df_train['SalePrice'])
```

```python
(mu,sigma) = norm.fit(df_train['SalePrice'])
print('mu = {:.2f} and sigma = {:.2f}'.format(mu,sigma))

sns.distplot(df_train['SalePrice'],fit = norm);
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f})'.format(mu,sigma)],loc = 'best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution');
```

## 2. Missing Values

```python
ntrain = df_train.shape[0]
ntest = df_test.shape[0]

df_y_train = df_train['SalePrice']

df_all_data = pd.concat([df_train,df_test]).reset_index(drop = True)
df_all_data.drop(['Id'],axis = 1,inplace = True)
#df_all_data.drop(['SalePrice'],axis = 1,inplace = True)

df_all_data.head()
```

```python
df_all_data.shape
```

```python
df_total = df_all_data.isnull().sum().sort_values(ascending = False)
df_percent = ((df_all_data.isnull().sum() / len(df_all_data)).sort_values(ascending = False)) * 100

df_missing_data = pd.concat([df_total,df_percent],axis = 1,keys = ['Total','Percent'])
df_missing_data = df_missing_data[df_missing_data['Total'] != 0]

df_missing_data
```

```python
plt.figure(figsize = (13,5))
sns.heatmap(df_all_data.isnull())
plt.show()
```

```python
plt.figure(figsize = (13,5))
sns.barplot(x = df_missing_data.index,y = 'Percent',data = df_missing_data)
plt.xticks(rotation = 90)
plt.show()
```

## Imputting Missing Values

```python
df_all_data['PoolQC'] = df_all_data['PoolQC'].fillna('None')
```

```python
df_all_data['MiscFeature'] = df_all_data['MiscFeature'].fillna('None')
```

```python
df_all_data['Alley'] = df_all_data['Alley'].fillna('None')
```

```python
df_all_data['Fence'] = df_all_data['Fence'].fillna('None')
```

```python
df_all_data['FireplaceQu'] = df_all_data['FireplaceQu'].fillna('None')
```

```python
df_all_data['LotFrontage'] = df_all_data.groupby('Neighborhood')['LotFrontage'].apply(lambda x: x.fillna(x.median()))
```

```python
for col in ['GarageType','GarageFinish','GarageQual','GarageCond']:
    df_all_data[col] = df_all_data[col].fillna('None')
```

```python
for col in ['GarageYrBlt', 'GarageArea', 'GarageCars']:
    df_all_data[col] = df_all_data[col].fillna(int(0))
```

```python
for col in ['BsmtFinType2','BsmtExposure','BsmtFinType1','BsmtCond','BsmtQual']:
    df_all_data[col] = df_all_data[col].fillna('None')
```

```python
df_all_data['MasVnrArea'] = df_all_data['MasVnrArea'].fillna(int(0))
```

```python
df_all_data['MasVnrType'] = df_all_data['MasVnrType'].fillna('None')
```

```python
df_all_data['Electrical'] = df_all_data['Electrical'].fillna(df_all_data['Electrical']).mode()[0]
```

```python
df_all_data = df_all_data.drop(['Utilities'],axis = 1)
```
