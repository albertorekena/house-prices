```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from scipy.stats import norm

from sklearn.model_selection import train_test_split, KFold, RepeatedKFold
from catboost import CatBoostRegressor, Pool, cv

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns',100)
```

```python
def run_catboost(df_train,df_test,df_hold,params,n_splits = 10,n_repeats = 1,plot = False,use_features = None,
                 plot_importance = True):
    folds = RepeatedKFold(n_splits = n_splits,n_repeats = n_repeats)
    p_hold = np.zeros(df_hold.shape[0])
    y_hold = df_hold.target
    p_test = np.zeros(df_test.shape[0])
    
    if use_features is None:
        use_features = df_test['columns'].values
    
    cat_features = np.where(df_test.loc[:,use_features].dtypes == 'object')[0]
    x_hold = df_hold.loc[:,use_features]
    x_test = df_test.loc[:,use_features]
    
    df_feature_importance = pd.DataFrame(index = use_features)
    
    m = 0
    cv_scores = []
    for train_idx,dev_idx in folds.split(df_train):
        x_train,x_dev = df_train.iloc[train_idx][use_features],df_train.iloc[dev_idx][use_features]
        y_train,y_dev = df_train.target.iloc[train_idx],df_train.target.iloc[dev_idx]
        train_pool = Pool(x_train,y_train,cat_features = cat_features)
        dev_pool = Pool(x_dev,y_dev,cat_features = cat_features)
        model = CatBoostRegressor(**params)
        model.fit(train_pool,eval_set = dev_pool,plot = plot)
        p_hold += model.predict(x_hold) / (n_splits*n_repeats)
        log_p_test = model.predict(x_test)
        p_test += (np.exp(log_p_test) - 1) / (n_splits*n_repeats)
        y_pred = model.predict(x_dev)
        df_feature_importance.loc[:,'fold_' + str(m)] = model.get_feature_importance(train_pool)
        cv_scores.append(np.sqrt(mse(y_dev,y_pred)))
        m += 1

    print("hold out rmse: " + str(np.sqrt(mse(y_hold,p_hold))))
    print("cv mean rmse: " + str(np.mean(cv_scores)))
    print("cv std rmse: " + str(np.std(cv_scores)))
    
    df_feature_importance['mean'] = df_feature_importance.mean(axis = 1)
    df_feature_importance['std'] = df_feature_importance.std(axis = 1)
    df_feature_importance = df_feature_importance.sort_values(by = 'mean',ascending = False)
    
    if plot_importance:
        plt.figure(figsize = (15,20))
        sns.barplot(x = df_feature_importance['mean'].values,y = df_feature_importance.index.values);
        plt.title('Feature importances');
        plt.show()
    
    results = {'last_model':model,
               'last_train_pool':train_pool,
               'feature_importance':df_feature_importance, 
               'p_hold':p_hold,
               'p_test':p_test,
               'cv_scores':cv_scores}
    return results
```

## Categorical Features

```python
cat_candidates = list(combined.dtypes[combined.dtypes == 'object'].index.values)

cat_candidates
```

## Numerical to Categorical Features

```python
num_to_cats = ['BsmtHalfBath','HalfBath','KitchenAbvGr','BsmtFullBath','Fireplaces','FullBath','GarageCars',
               'BedroomAbvGr','OverallCond','OverallQual','TotRmsAbvGrd','MSSubClass','YrSold','MoSold', 
               'GarageYrBlt','YearRemodAdd']

for feat in num_to_cats:
    num_candidates.remove(feat)
    cat_candidates.append(feat)
    combined[feat] = combined[feat].astype('object')
    df_train[feat] = df_train[feat].astype('object')
    df_test[feat] = df_test[feat].astype('object')
```

## Useless Categorical Features

```python
frequencies = []

for col in cat_candidates:
    overall_freq = combined.loc[:,col].value_counts().max() / combined.shape[0]
    frequencies.append([col,overall_freq])

frequencies = np.array(frequencies)
freq_df = pd.DataFrame(index=frequencies[:,0], data=frequencies[:,1], columns=["frequency"])
sorted_freq = freq_df.frequency.sort_values(ascending=False)

plt.figure(figsize=(20,5))
sns.barplot(x=sorted_freq.index[0:30], y=sorted_freq[0:30].astype(np.float), palette="Blues_r")
plt.xticks(rotation=90);
```
